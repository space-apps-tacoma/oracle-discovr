```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
first_column_data = data.iloc[:, 1]
second_column_data = data.iloc[:, 2]
third_column_data = data.iloc[:, 3]
data['magnitude'] = ((first_column_data ** 2) + (second_column_data ** 2) + (third_column_data ** 2))**0.5
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.xlim(-20, 15)
plt.ylim(-20, 20)
ax.scatter(first_column_data, second_column_data, third_column_data)
plt.show()
```

```python
date = data.iloc[:, 0]
plt.plot(date, first_column_data)
plt.show()
```
