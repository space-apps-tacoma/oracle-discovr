```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('dsc_fc_summed_spectra_2016_v01.csv', header=None)
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data[0]= pd.to_datetime(data[0])
```

```python
data.replace(0, np.nan, inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
Bx = data.iloc[:, 1]
By = data.iloc[:, 2]
Bz = data.iloc[:, 3]
data['magnitude'] = ((Bx ** 2) + (By ** 2) + (Bz ** 2))**0.5
```

```python
data.head()
```

```python
# Figure of Bx, By, Bz
# Matplotlib does not accomodate for third axes (z) but just note that the third axes is for 'Bz'
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.xlim(-20, 15)
plt.ylim(-20, 20)
plt.title('Figure of Bx, By, Bz')
plt.xlabel('Bx (nT)')
plt.ylabel('By (nT)')
ax.scatter(Bx, By, Bz)
plt.show()
```

```python
# Figure of Bx with time
date = data.iloc[:, 0]
plt.title('Change in Bx with time')
plt.xlabel('time')
plt.ylabel('Bx (nT)')
plt.plot(date, Bx)
plt.show()
```

```python
# Figure of By with time
date = data.iloc[:, 0]
plt.title('Change in By with time')
plt.xlabel('time')
plt.ylabel('By (nT)')
plt.plot(date, By)
plt.show()
```

```python
# Figure of Bz with time
date = data.iloc[:, 0]
plt.plot(date, Bz)
plt.title('Change in Bz with time')
plt.xlabel('time')
plt.ylabel('Bz (nT)')
plt.show()
```

```python
# Figure of magnitude (Bt) with time
date = data.iloc[:, 0]
plt.plot(date, data['magnitude'])
plt.title('Change in Magnitude (Bt) with time')
plt.xlabel('time')
plt.ylabel('Magnitude')
plt.show()
```
