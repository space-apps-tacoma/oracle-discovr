```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_2016 = pd.read_csv('dsc_fc_summed_spectra_2016_v01.csv', header=None)
data_2017 = pd.read_csv('dsc_fc_summed_spectra_2017_v01.csv', header=None)
data_2018 = pd.read_csv('dsc_fc_summed_spectra_2018_v01.csv', header=None)
data_2019 = pd.read_csv('dsc_fc_summed_spectra_2019_v01.csv', header=None)
data_2020 = pd.read_csv('dsc_fc_summed_spectra_2020_v01.csv', header=None)
data_2021 = pd.read_csv('dsc_fc_summed_spectra_2021_v01.csv', header=None)
data_2022 = pd.read_csv('dsc_fc_summed_spectra_2022_v01.csv', header=None)
data_2023 = pd.read_csv('dsc_fc_summed_spectra_2023_v01.csv', header=None)
```

```python
data = [data_2016, data_2017, data_2018, data_2019, data_2020, data_2021, data_2022, data_2023]
```

```python
df = pd.concat(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df[0]= pd.to_datetime(df[0])
```

```python
df.replace(0, np.nan, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
Bx = df.iloc[:, 1]
By = df.iloc[:, 2]
Bz = df.iloc[:, 3]
df['magnitude'] = ((Bx ** 2) + (By ** 2) + (Bz ** 2))**0.5
```

```python
df.head()
```

```python
# Figure of Bx, By, Bz
# Matplotlib does not accomodate for third axes (z) but just note that the third axes is for 'Bz'
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.xlim(-20, 15)
plt.ylim(-20, 20)
plt.title('Figure of Bx, By, Bz')
plt.xlabel('Bx (nT)')
plt.ylabel('By (nT)')
ax.scatter(Bx, By, Bz)
plt.show()
```

```python
# Figure of Bx with time
date = df.iloc[:, 0]
plt.title('Change in Bx with time')
plt.xlabel('time')
plt.ylabel('Bx (nT)')
plt.plot(date, Bx)
plt.show()
```

```python
# Figure of By with time
date = df.iloc[:, 0]
plt.title('Change in By with time')
plt.xlabel('time')
plt.ylabel('By (nT)')
plt.plot(date, By)
plt.show()
```

```python
# Figure of Bz with time
date = df.iloc[:, 0]
plt.plot(date, Bz)
plt.title('Change in Bz with time')
plt.xlabel('time')
plt.ylabel('Bz (nT)')
plt.show()
```

```python
# Figure of magnitude (Bt) with time
date = df.iloc[:, 0]
plt.plot(date, df['magnitude'])
plt.title('Change in Magnitude (Bt) with time')
plt.xlabel('time')
plt.ylabel('Magnitude')
plt.show()
```

```python
df.to_csv('all_experimental.csv', index=False)
```
